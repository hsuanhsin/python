```python
import os
import string
import re
import pandas
import codecs
import glob
import pandas as pd
import json
import numpy as np
os.getcwd()
os.chdir('/Volumes/kiki/VTReport')
list = [] #原本的filename包含txt
list2 = []#去掉txt的病毒編號
list3 = []#全部的判斷名字字串
list4 = [[0 for i in range(79)] for j in range(1000)]#字串去掉標點符號
name_list = []#廠商名
files = glob.glob('*.txt')
f = glob.iglob('*.txt')


for filename in f:
    list.append(filename)
for name in list:
    list2.append(name.strip('.txt'))
    
#print(list2)
for i in range(0, 1000):
    all = open(list[i])
    data = json.load(all)
    data2 = data['scans']
    #print(data2)
    df = pd.DataFrame(data2)
    df1 = df.values
    df2 = df1[1].tolist()
    #print(data2)
    list3.append(df2)   
    for col in df.columns:
        if col not in name_list:
            name_list.append(col)
#遇到的問題是不同編號的病毒廠商數量可能不同！
# 把廠商先從資料中取出，廠商樣本缺乏的都補None
   

#print(list3)
#把標點符號用空格代替
num = 0
index = 0
while(num != 1000 ):
    for name in list3[num]:
        if name != None:
            for word in name:
                 if word in string.punctuation:
                    name = name.replace(word," ") 
            #print(name)
            list4[num][index] = name
            index = index +1
        else:
            #print(name)
            list4[num][index] = None
            index = index +1
    index = 0
    num = num +1
    
```

```python
rdf =pd.DataFrame.from_records(list4,columns=name_list, index=list2[0:1000])
rdf1 = rdf.replace({0:None})
a =rdf1.isnull()
b = [] #把廠商判斷none高於80%的名字都存此

for name in a.columns:
    #print(sum(a[name]))
    if sum(a[name]) > 800:
        #print(name)
        b.append(name)
rdf2 = rdf1.drop(b,axis = 1)#把判斷很差的廠商column都drop掉
rdf3 = rdf2.dropna(how ="all",axis = 0)

```

```python
list_gen = ["adware", "agent", "application", "artemis", "appl", "bundler", 
            "behaveslike", "backdoor", "crypt", "cryptor", "downloader", "dropper", 
            "downloadware", "eldorado", "emailworm", "fakealert", "gen", "generic", 
            "heur", "heuristic", "injector", "malware","obfuscator","pup", "suspicious",
            "trj", "troj", "trojan", "virus", "autorun", "unwanted", "variant", "virus",
            "win32", "worm", "w32", "win","malicious","soft32downloader"]
gt = []
gt_count=[]


def is_int(str):
    try:
        int(str)
        return True
    except ValueError:
        return False
num = 0
print(list4[44])
while num < 100:
    row_freq = {}
    for freq in list4[num]:
        if freq is None:
            continue
        if freq == 0:
            continue
        freq_name = freq.lower().split(" ")
        for t in freq_name:
            if len(t)>= 3 and not is_int(t):
                    #print(t)
                if t not in row_freq:
                    row_freq[t] = 1
                else:
                    row_freq[t]+= 1

    for key in row_freq.copy():
        if key in list_gen:
            row_freq.pop(key)

    count = 0
    for w in sorted(row_freq, key=row_freq.get, reverse=True):
        #print(w, row_freq[w])
            
        if count < 3:
            print(w, row_freq[w])
            gt.append(w)
            gt_count.append(row_freq[w])
            count = count+1
    while count != 3:
        gt.append(None)
        gt_count.append(0)
        count = count+1
   
        
    num = num + 1
print(gt)
print(gt_count)
```
